# Matplot lib - examples of plots

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('data0.csv')
```

## Plot with lines

```python
plt.plot(df['x'],df['y'])
```

## Plot with points

```python
plt.scatter(df['x'],df['y'])
```

## Adding text and lines to the plot

```python
ax = plt.subplot(111)
plt.scatter(df['x'],df['y'])
plt.annotate('no points here', xy=(750, 350), xytext=(1000, 200),
             arrowprops=dict(facecolor='black', shrink=0.05),
             )
plt.axhline(250,0,1, color='black')
plt.axhline(400,0,1, color='black')

plt.show()
```

## Bar plot - for each pair (x,y)

```python
plt.bar(df['x'],df['y'])
```

## Histogram - by default 10 bins

```python
plt.hist(df['x'])
```

## Histogram with four bins

```python
plt.hist(df['y'],bins=4)
```
